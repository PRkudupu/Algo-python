You are given two strings s and t.

String t is generated by random shuffling string s and then add one more letter at a random position.

Return the letter that was added to t.

 

<b>Example 1:</b>

Input: s = "abcd", t = "abcde"<br>
Output: "e"<br>
Explanation: 'e' is the letter that was added.<br>

<b>Example 2:</b>

Input: s = "", t = "y"<br>
Output: "y"<br>

<b>Example 3:</b>

Input: s = "a", t = "aa"<br>
Output: "a"<br>

<b>Example 4:</b>

Input: s = "ae", t = "aea"<br>
Output: "a"<br>
 

Constraints:<br>

0 <= s.length <= 1000<br>
t.length == s.length + 1<br>
s and t consist of lower-case English letters.<br>

```python
def find_the_difference(s,t):
    dic ={}
    for i in s:
        #increment if the key is found
        if i in dic:
            dic[i] +=1
        else:
            dic[i]=1
    for j in t:
        if j not in dic:
            return j
        #If the key is 0 then return
        elif dic[j]==0:
            return j
        else:
            dic[j]-=1
```

```python
print(find_the_difference("abcd","abcde"))
print(find_the_difference("a","aa"))
print(find_the_difference("ae","aea"))
```

Complexity Analysis

<b>Time Complexity: O(N),</b> where N is length of the strings. Since, we iterate through both the strings once.

<b>Space Complexity: O(1).</b> The problem states string s and string t have lowercase letters. Thus, the total number of unique characters and eventually buckets in the hash map possible are just 26.



```python
#bruteforce method
#Note one of the test case fails in leetcode
def find_the_difference(s,t):
    dic = {}
    for i in s:
        dic[i]=1
    for j in t:
        if j in dic.keys() and dic[j]==1:
            dic[j]=0
        else:
            dic[j]=1
    for key,values in dic.items():
        if values==1:
            return key
```

```python
print(find_the_difference("abcd","abcde"))
print(find_the_difference("","y"))
print(find_the_difference("a","aa"))
print(find_the_difference("ae","aea"))
```
